```python
import re 
import xlwt
from xlwt import Workbook

# Workbook is created
wb = Workbook()

# add_sheet is used to create sheet.
sheet1 = wb.add_sheet('Sheet 1')

# Add sheet Heading name 
#sheet1 .write('A1', 'Credit Card no.')
#sheet1 .write('B1', 'result')

# open text file 
with open("Credit Card Nos.txt") as f:
    Number = f.readlines()
    
    # use Exception Handling and Regular expressions
    try:
        assert re.match(r'(\d{4}-){3}\d{4}$', Number) or re.match(r'\d{16}$', Number)
        Number = re.sub(r'-', '', Number)
        assert re.match(r'[456]', Number)
        assert not re.search(r'(\d)\1{3,}', Number)
    except:
        result = 'Invalid'
    else:
        result ='Valid'

    sheet1.write(Number, result)
    # save the result

wb.save('Credit_Card_no_test.xlsx')


    
     
    
    
```
